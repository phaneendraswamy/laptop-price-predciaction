```python
df = df

# Define the distribution rule for high-res values
high_res = ['4K Ultra HD', 'Quad HD+', 'WQHD+']
low_res = ['Full HD', 'HD Ready', 'QHD']

# Create a mapping of high-res to low-res categories
distribution_map = {
    '4K Ultra HD': 'Full HD',  # Assign to Full HD
    'Quad HD+': 'QHD',         # Assign to QHD
    'WQHD+': 'QHD',           # Assign to QHD
}

# Replace high-res values based on distribution map
df['ScreenResolution'] = df['ScreenResolution'].replace(distribution_map)

# For better distribution, you could randomly assign from `low_res`
# Example of random distribution of high-res values to low-res categories
np.random.seed(0)  # For reproducibility
for index, row in df.iterrows():
    if row['ScreenResolution'] in high_res:
        df.at[index, 'ScreenResolution'] = np.random.choice(low_res)

# Display the updated DataFrame
print(df)
```
